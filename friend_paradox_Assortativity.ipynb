```
import networkx as nx
import pandas as pd
import statistics
import seaborn as sns
import numpy as np
```

```
def compute_neighbors_avg_degree(aGraph):
  node_id=[]
  degree=[]
  neighbors_degrees=[]

  for n in aGraph.node():

    neighbors = aGraph[n]
    if len(neighbors)>0:
      node_id.append(n)

      degree.append(len(neighbors))

      degrees = nx.degree(aGraph,neighbors)
      neighbors_degrees.append(statistics.mean(dict(degrees).values()))
  stats = pd.DataFrame()
  stats["ids"]=node_id
  stats["degree"]=degree
  stats["neighb_avg_degree"]=neighbors_degrees
  return stats
```

```
def analyse_a_network(net):
  stat_deg = compute_neighbors_avg_degree(net)
  print("avg_degree = ",statistics.mean(stat_deg["degree"]))
  print("avg_neighbors_degree_by_node = ",statistics.mean(stat_deg["neighb_avg_degree"]))
  print("avg_neighbors_degree_overall = ",np.average(stat_deg["neighb_avg_degree"],weights=stat_deg["degree"]))
  print("degree_assortativity = ",nx.degree_assortativity_coefficient(net))

  sns.lineplot(x="degree",y="neighb_avg_degree",data=stat_deg)
```

```
ER_g = nx.erdos_renyi_graph(10000,0.001)
analyse_a_network(ER_g)
```

```
WS = nx.watts_strogatz_graph(10000,6,0.3)
analyse_a_network(WS)
```

```

config_PL = nx.expected_degree_graph(nx.utils.powerlaw_sequence(10000, 2.2))
analyse_a_network(config_PL)
```
